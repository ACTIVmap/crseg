```python
import networkx as nx
import osmnx as ox

```

Chargement des données de Clermont-Ferrand avec simplification de la géométrie (on garde les sommets qui ne sont pas des croisements), puis affichage

```python
G = ox.graph_from_bbox(north=45.77811, west=3.07116, south=45.77644, east=3.07537, network_type="drive", retain_all=True, truncate_by_edge=True, simplify=False)
fig, ax = ox.plot_graph(G)
```

Quelques exemples de références pour comprendre comment tout cela fonctionne:
* [Coloring nodes according to an attribute in OSMnx](https://stackoverflow.com/questions/62817760/coloring-nodes-according-to-an-attribute-in-osmnx)
* [OSMnx: Python for Street Networks](https://deeplearning.lipingyang.org/wp-content/uploads/2019/03/OSMnx_-Python-for-Street-Networks-Geoff-Boeing.pdf), avec de la documentation sur la manière de colorier les segments


```python
ec = ox.plot.get_edge_colors_by_attr(G, attr='length')
ox.plot.plot_graph(G, edge_color=ec)
```
